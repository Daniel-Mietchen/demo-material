```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# bigger range
xx1 = np.linspace(-0.3, .1)

# actual range for the parking maneuver
xx2 = np.linspace(-0.2, .0)


def park_poly(xx):
    return -22.5*xx**2 -75*xx**3
 
plt.figure(figsize=(10, 10))
plt.axis('equal')  # fix aspect ratio to 1:1

plt.plot(xx1, park_poly(xx1), "--", color="blue", lw=0.5)
plt.plot(xx2, park_poly(xx2), "-", color="blue", lw=2)
plt.grid(True)
plt.xlabel("$x$ in m")
plt.ylabel("$y$ in m")
```
