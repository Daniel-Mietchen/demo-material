```python
from graphviz import Digraph

# adapted from example: https://graphviz.readthedocs.io/en/stable/examples.html
```

```python
f = Digraph('knowledge_base', filename='kb.gv')
f.attr(rankdir='TB', size='8,5')

def node(graph, *args, **kwargs):
    graph.node(*args, **kwargs)
    return args[0]

def edge(graph, *args, **kwargs):
    graph.edge(*args, **kwargs)
    return args[1]

n1 = node(f, "allgemeines lin.\n DGL-System")
n2 = edge(f, n1, "lin. Zustandsraum-\ndarstellung")
n3 = edge(f, n2, "Übertragungs-\n funktion", "Regelungs-\n normalform")
edge(f, n1, n3, "Laplace-Transformation")
n4 = edge(f, n3, "Charakteristisches\n Polynom")
n5 = edge(f, n4, "Stabilitätsaussage", "Alle Polstellen links?")
n6 = edge(f, n3, "Basisgrößen\n (flacher Ausgang)")
n7 = edge(f, n2, "Eigenwerte d.\n Systemmatrix")
n8 = edge(f, n7, n5, "Alle EW\n links?")
n9 = node(f, "Anfangs- und\n Endwerte")
n10 = edge(f, n9, "Trajektorienplanung")
edge(f, n6, n10)



f # implicitly calls f._repr_svg_()

# f.view() # create pdf-file
```
